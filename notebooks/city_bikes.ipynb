```python
import requests

# Function to retrieve bike station data for Hamilton from CityBikes API
def get_hamilton_bike_stations():
    api_url = "https://api.citybik.es/v2/networks/sobi-hamilton"
    response = requests.get(api_url)
    if response.status_code == 200:
        data = response.json()
        stations = data["network"]["stations"]
        return stations
    else:
        print("Failed to retrieve data from CityBikes API.")
        return None

# Retrieve bike station data for Hamilton
hamilton_stations = get_hamilton_bike_stations()

# Print the bike station data for Hamilton
if hamilton_stations is not None:
    print("Bike Stations in Hamilton:")
    for station in hamilton_stations:
        print("- " + station["name"])
else:
    print("No bike station data available for Hamilton.")
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
import requests

# Define the API endpoint URL to retrieve data for the SOBI Hamilton network
url = "http://api.citybik.es/v2/networks/sobi-hamilton"

# Make the GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract station details
    stations = data['network']['stations']
    
    # Function to parse the response and extract details for each bike station
    for station in stations:
        # Extract station name, latitude, longitude, and number of bikes
        name = station['name']
        latitude = station['latitude']
        longitude = station['longitude']
        bikes_available = station['free_bikes']
        
        # Print station details
        print(f"Station: {name}, Latitude: {latitude}, Longitude: {longitude}, Bikes available: {bikes_available}")
else:
    # Handle errors
    print(f'Error: {response.status_code} - {response.text}')
```

Put your parsed results into a DataFrame.

```python
import requests
import pandas as pd

# Define the API endpoint URL to retrieve data for the SOBI Hamilton network
url = "http://api.citybik.es/v2/networks/sobi-hamilton"

# Make the GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract station details
    stations = data['network']['stations']
    
    # Create lists to store parsed data
    station_names = []
    latitudes = []
    longitudes = []
    bikes_available = []
    
    # Parse data and store in lists
    for station in stations:
        station_names.append(station['name'])
        latitudes.append(station['latitude'])
        longitudes.append(station['longitude'])
        bikes_available.append(station['free_bikes'])
    
    # Create DataFrame from parsed data
    df = pd.DataFrame({
        'Station Name': station_names,
        'Latitude': latitudes,
        'Longitude': longitudes,
        'Bikes Available': bikes_available
    })
    
    # Print DataFrame
    print(df)
else:
    # Handle errors
    print(f'Error: {response.status_code} - {response.text}')
```

```python
df.to_csv('..\\data\\available_bikes.csv', index=False)
```
